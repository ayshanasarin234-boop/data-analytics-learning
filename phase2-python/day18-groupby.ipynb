```python
import pandas as pd
```

```python
data = {
    "OrderID": [101,102,103,104,105,106,107,108],
    "Product": ["Laptop","Phone","Tablet","Shoes","Shoes","Watch","Phone","Laptop"],
    "Category": ["Electronics","Electronics","Electronics","Fashion","Fashion","Accessories","Electronics","Electronics"],
    "City": ["Delhi","Mumbai","Delhi","Bangalore","Bangalore","Chennai","Delhi","Mumbai"],
    "Sales": [55000,30000,20000,4000,4000,5000,30000,55000],
    "Quantity": [1,2,1,2,2,1,2,1]
}

df = pd.DataFrame(data)
print(df)
```

```python
df.describe()
```

```python
df.info()
```

```python
total_sales = df.groupby("Category")["Sales"].sum()
print(total_sales)
# Grouped data by Category and calculated total Sales using sum()
```

```python
avg_sales = df.groupby("Category")["Sales"].mean()
print(avg_sales)
# Calculated average Sales for each Category using mean()
```

```python
count_rows=df.groupby("Category")["Sales"].count()
print(count_rows)
# Counted number of records in each Category
```

```python
df["Category"].value_counts()
```

```python
df.groupby("Category")["Sales"].agg(["sum","mean","count"])
```

# Comparison with Power BI

The grouped results in Pandas match the values shown in my Power BI bar chart.

Both tools calculated total sales per category correctly.

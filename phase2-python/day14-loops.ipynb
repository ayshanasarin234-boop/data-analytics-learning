```python
scores=[75,35,20,30,50,55,67,25,80,28]
print("student scores:",scores)
```

```python
for score in scores:
    if score >= 50:
        print(score,"- pass")
    else:
        print(score,"- fail")
```

```python
pass_count=0
fail_count=0
for score in scores:
    if score >= 50:
        pass_count += 1
    else:
        fail_count += 1
print("passed:",pass_count)
print("failed:",fail_count)
```

```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv(r"C:\Users\AYSHA NASARIN\OneDrive\Documents\day20_dataset.csv")
```

```python
df.info()
```

```python
category_sales = df.groupby("Category")["Sales"].sum()
plt.figure(figsize=(8,4))
plt.bar(category_sales.index,category_sales.values,color="blue")
plt.title("Total Sales by Category")
plt.xlabel("Category")
plt.ylabel("Sales")
plt.show()
```

This bar chart shows total profit earned from each product category

```python
plt.figure(figsize=(8,4))
plt.plot(df["Order_Date"],df["Sales"],color="red")
plt.title("Sales Trend by Date")
plt.xlabel("Index")
plt.ylabel("Sales")
plt.show()
```

This line chart shows how sales values change across the dataset.

```python
plt.figure(figsize=(8,4))
plt.hist(df["Sales"], bins=20,color="green")
plt.title("Sales Distribution")
plt.xlabel("Sales")
plt.ylabel("Frequency")
plt.show()
```

This histogram shows the distribution of sales values and where most values are concentrated

```python
category_sales = df.groupby("Category")["Sales"].sum()
plt.figure(figsize=(8,4))
plt.bar(category_sales.index,category_sales.values,color="Orange")
plt.title("Sales by Category (Recreated from Power BI")
plt.xlabel("Category")
plt.ylabel("Sales")
plt.show()
```

This chart recreates the Power BI visualization showing total sales across categories.

```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv(r"C:\Users\AYSHA NASARIN\OneDrive\Documents\day20_dataset.csv")
```

```python
x=df["Sales"]
y=df["Profit"]
plt.scatter(x, y)
plt.title("Sales vs Profit")
plt.xlabel("Sales")
plt.ylabel("Profit")
plt.show()
```

Scatter Plot Observation

The two variables (Sales and Profit) do not show a strong relationship. The data points are widely scattered, indicating a weak or no clear correlation. As sales increase, profit does not follow a consistent pattern—some high sales values correspond to low or even negative profit. The pattern appears random rather than linear, with noticeable variability and a few outliers.

```python
plt.hist(df["Sales"], bins=10)
plt.title("Sales Distribution")
plt.xlabel("Sales")
plt.ylabel("Frequency")
plt.show()
```

Histogram Observation

The distribution of Sales is spread out across a wide range, indicating high variability rather than concentration in a narrow range. The values are distributed unevenly, with slightly more observations in the mid to higher range. There are no extreme outliers, but a few higher values near the upper limit stand apart from the rest.

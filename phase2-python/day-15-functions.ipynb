```python
def celsius_to_farenheit(temp): 
    return (9/5) * temp + 32

print(celsius_to_farenheit(0))
print(celsius_to_farenheit(10))
print(celsius_to_farenheit(100))
print(celsius_to_farenheit(50))
```

```python
def get_grade(score):
    if score >= 90:
        return "A"
    elif score >= 75:
        return "B"
    elif score >= 50:
        return "C"
    else:
        return "F"

print(get_grade(95))  
print(get_grade(82))  
print(get_grade(67))  
print(get_grade(45))  
print(get_grade(50))  
```

This function takes a temperature in Celsius as input.
It converts the temperature to Fahrenheit using a formula.
It then returns the converted value.

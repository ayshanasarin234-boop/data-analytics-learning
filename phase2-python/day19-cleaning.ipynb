```python
import pandas as pd
```

```python
df=pd.read_csv(r"C:\Users\AYSHA NASARIN\Desktop\day3_power.csv")
df.shape
```

```python
df.info()
```

```python
# Find missing values
# Reason: To identify which columns contain null values
df.isnull().sum()
```

```python
# Convert Order_Date to datetime
# Reason: To ensure proper date format for analysis

df["Order_Date"] = pd.to_datetime(df["Order_Date"], errors="coerce")
```

```python
df.info()
```

```python
print(df)
```

```python
# Handle missing dates using forward fill
# Reason: Preserve rows instead of dropping; assumes data is in order
df["Order_Date"] = df["Order_Date"].fillna(method="ffill")
```

```python
df.info()
```

```python
print(df)
```

```python
# Drop rows where Product is missing
# Reason: Product is a key column required for analysis
df = df.dropna(subset=["Product"])
```

```python
# Fill missing Customer_Name
# Reason: Only few missing values; filling avoids data loss
df["Customer_Name"]=df["Customer_Name"].fillna("Unknown")
print(df)
```

```python
#  Convert Price to float
# Reason: Ensure numeric operations can be performed
df["Price"] = df["Price"].astype(float)
```

```python
# Remove duplicate rows
# Reason: Duplicate data can distort analysis results
df = df.drop_duplicates()
```
